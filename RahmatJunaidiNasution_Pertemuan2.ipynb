```python
def totallist(a, b):
  total = 0
  for i in a:
    total += i
  for i in b:
    total += i
  return total

a = [1, 2, 3]
b = [4, 5, 6]

result = totallist(a, b)
print(result)
```

```python
# prompt: getsum getkali list=1,3,5,7

def getsum(list):
  total = 0
  for i in list:
    total += i
  return total

def getkali(list):
  total = 1
  for i in list:
    total *= i
  return total

list = [1, 3, 5, 7]
print(getsum(list))
print(getkali(list))
```
