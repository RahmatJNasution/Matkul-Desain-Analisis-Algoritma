```python


vektor_a = [1, 2, 3]
vektor_b = [4, 5, 6]

vektor_c = [a + b for a, b in zip(vektor_a, vektor_b)]
print("Penjumlahan vektor:", vektor_c)

vektor_d = [a - b for a, b in zip(vektor_a, vektor_b)]
print("Pengurangan vektor:", vektor_d)

skalar = 2
vektor_e = [a * skalar for a in vektor_a]
print("Perkalian skalar:", vektor_e)
```
