```python
def cari_angka_terbesar(angka1, angka2):
  """
  Fungsi untuk mencari angka terbesar dari dua angka.
asd
  Args:
    angka1: Angka pertama.
    angka2: Angka kedua.

  Returns:
    Angka terbesar.
  """
  if angka1 > angka2:
    return angka1
  else:
    return angka2


angka1 = float(input("Masukkan angka pertama: "))
angka2 = float(input("Masukkan angka kedua: "))

angka_terbesar = cari_angka_terbesar(angka1, angka2)
print("Angka terbesar adalah:", angka_terbesar)
```
